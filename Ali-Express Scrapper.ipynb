```python
import requests
import pandas as pd # data transformation
from bs4 import BeautifulSoup
import csv
import json
```

```python
for page_num in range(1,61):
    ROOT_URL = f"https://www.aliexpress.com/w/wholesale-Baggy-Pants.html?isFromCategory=y&categoryUrlParams=%7B%22q%22%3A%22Baggy+Pants%22%2C%22s%22%3A%22qp_nw%22%2C%22osf%22%3A%22category_navigate%22%2C%22sg_search_params%22%3A%22%22%2C%22guide_trace%22%3A%222ee73bb2-4e3f-4e2d-88a7-fd15a74c156f%22%2C%22scene_id%22%3A%2237749%22%2C%22searchBizScene%22%3A%22openSearch%22%2C%22recog_lang%22%3A%22en%22%2C%22bizScene%22%3A%22category_navigate%22%2C%22guideModule%22%3A%22category_navigate_vertical%22%2C%22postCatIds%22%3A%22200000343%2C200001813%22%2C%22scene%22%3A%22category_navigate%22%7D&page={page_num}&g=y&SearchText=Baggy+Pants"
    product_data = [] #empty list
    
#lets make a request    
def make_request(page_num:int, page_url=ROOT_URL):
    try:
        response = requests.get(url=page_url)
        if response.status_code == 200:
            content = response.content
            return content
    except:
        print("Invalid response")
        
        
def save_to_csv(filename, products):
    #save to csv
    with open(filename, mode="w", encoding="utf-8")as csv_file:
        if products:
            FIELDNAMES = products[0].keys()
            csv_writer = csv.DictWriter(csv_file, fieldnames=FIELDNAMES)
            csv_writer.writeheader()
            for product in products:
                csv_writer.writerow(product)
        print("Done writing to CSV")
        
def save_to_json(filename,products):
    #save to json
    with open(filename, mode="w")as json_file:
        json.dump(products, json_file)
    print("Done writing to json")
        
def fetch_prod_info(product:str) -> dict:
    
    
    # product name
    product_name = product.find("div", class_="multi--title--G7dOCj3")
    if product_name != None:
        product_name = product_name.text
    else:
        product_name = ""
        
    
    # qty_sold
    qty_sold = product.find("span", class_="multi--trade--Ktbl2jB")
    if qty_sold != None:
        qty_sold  = qty_sold.text
    else:
        qty_sold = "" 
        
    
    # sales_price
    sales_price = product.find("div", class_="multi--price-sale--U-S0jtj")
    if sales_price != None:
        sales_price  = sales_price.text
    else:
        sales_price = ""
        
        
    # original_price
    original_price = product.find("div", class_="multi--price-original--1zEQqOK")
    if original_price != None:
        original_price  = original_price.text
    else:
        original_price = ""
        
        
    # rating 
    rating = product.find("div", class_="multi--starList--t9_CAj2")
    if rating != None:
        rating  = rating.text
    else:
        rating = ""
        
        
    # shipping_cost
    shipping_cost = product.find("div", class_="multi--serviceContainer--3vRdzWN")
    if shipping_cost != None:
        shipping_cost  = shipping_cost.text
    else:
        shipping_cost = ""
        
        
    # store
    store = product.find("span", class_="cards--store--3GyJcot")
    if store != None:
        store = store.text
    else:
        store = ""
        
        
    # wrap product info into a dictionary
    product_info = {
        "product_name": product_name,
        "qty_sold": qty_sold,
        "sales_price": sales_price,
        "original_price": original_price,
        "rating": rating,
        "shipping_cost": shipping_cost,
        "store": store}
    
    return product_info

if __name__ == "__main__":
    for num in range (1,61):
        # make request
        try:
            content = make_request(page_num=num)
            print(f"done making a request to page {num}")
        except:
            print(f"invalid content for page {num}")
                  
        #create a soup 
        soup = BeautifulSoup(content, "html.parser")
        prod_container = soup.find_all("div", class_="list--gallery--C2f2tvm search-item-card-wrapper-gallery")
        
        for product in prod_container:
            #fetch prod_info
            product_info = fetch_prod_info(product)
            
            #append prod_info to product_data
            product_data.append(product_info)
        
        


        
        #save to csv
        save_to_csv(filename="aliexpress_product.csv", products=product_data)
        #save to json
        save_to_json(filename="aliexpress_product.json", products=product_data)
```
